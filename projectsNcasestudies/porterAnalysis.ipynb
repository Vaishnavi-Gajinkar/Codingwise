```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('porter clean data.csv')
df.describe().T
```

```python
df.describe(include='object').T
```

Observations:<br>
Most of the restaurants (9% overall) have 'american' dishes as speciality<br>

```python
plt.Figure(figsize=(12,8))
corr_matrix = df[['market_id','total_items','subtotal','num_distinct_items','min_item_price','max_item_price','week_number']].corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
