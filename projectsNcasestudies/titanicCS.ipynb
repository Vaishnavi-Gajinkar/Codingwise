```python
# Q10: Display details of passengers who were under 30 years old
df[df['age']<30]
```

```python
# Q11. Identify passengers who embarked from Queenstown (Q) or Cherbourg (C) and whose ticket class was either first or second class, but did not survive.
df[((df['embark_town']=='Queenstown') | (df['embark_town']=='Cherbourg')) & ((df['class']=='First') | (df['class']=='Second'))]
```

```python
# Q12: List all passengers who had a cabin on deck 'B' and were traveling alone (no siblings/spouses/parents/children).

df[(df['deck']=='B') & ((df['sibsp']==0) | (df['parch']==0))].head()
```

```python
# Q13. Find passengers who were older than the median age of the dataset, traveled in third class,
# and paid a fare less than the median fare of third-class passengers.

med_age = df['age'].median()
med_fare_3C = df.groupby(['class'])['fare'].median()['Third']
print(f'median age {med_age} \nmedian fare {med_fare_3C}')

df[(df['age']>med_age) & (df['class']=='Third') & (df['fare']<med_fare_3C)].head()
```

```python
# Q14: Show passengers who survived and are either staff (not having a passenger class-it means pclass is na) or children under 15 who were in any class.
# Assuming 'Staff' can be deduced by NaNs in 'Pclass' or some other criteria

print(f'Distinct items in pclass {df['pclass'].unique()}')
df[(df['survived']==1) & ( (df['pclass'].isna()) | ((df['who']=='child') & (df['age']<15) ))].head()
```

```python
# Q15: Display all female passengers from first class whose fare was above the average fare of first-class passengers and who embarked at Cherbourg (C).
avg_fare = df.groupby('class')['fare'].mean()['First']
print(f'Avg fare of 1st class is {avg_fare}')
df[(df['sex']=='female') & (df['class']=='first') & (df['fare']>avg_fare) & (df['embark_town']=='Cherbourg')]
```

```python
# Q16. Group passengers by class and find the average age and median fare for each class.

df.groupby('class').agg({'age':'mean','fare':'median'})
```

```python
# Q17 Sort the passengers by fare in descending order and display the top 10 most expensive tickets.

df.sort_values(by='fare',ascending=False).head(10)
```

```python
# Q18 Create a pivot table showing the average fare and survival rate grouped by passenger class and sex.

pivot_table = df.pivot_table(values=['fare', 'survived'],
                                  index=['pclass', 'sex'],
                                  aggfunc={'fare': 'mean', 'survived': 'mean'})
pivot_table
```

```python
df.pivot_table(values = ["fare", "age"],
               index = ["pclass", "sex"],
               aggfunc = {"fare" : 'mean', "age": "mean"})
```

```python
df.pivot_table(values = ["age"],
               index = ["pclass", "sex"],
               aggfunc = {"age": "mean"},
               fill_value = 29.6)
```

```python
# Q19: List the total number of survivors and non-survivors for each embarkation point.

survival_counts = df.groupby('embarked')['survived'].value_counts().unstack()
survival_counts
```

```python
# Q20: Find the top three ages with the highest average fare, grouped by passenger class.

df.groupby(['pclass','age'])['fare'].mean().sort_values(ascending=False).groupby('pclass').head(3)
```
