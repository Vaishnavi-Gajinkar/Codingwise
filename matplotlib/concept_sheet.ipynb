```python
# importing the library in this file
import matplotlib.pyplot as plt
```

# Line Graph

```python
x_vals = [1,2,3,4]
y_vals = [10,20,15,25]

plt.plot(x_vals, y_vals)
plt.show()
```

```python
# adding the chart details
weekdays = ['Mon','Tue','Wed','Thurs','Fri']
daily_sales = [10,15,7,20,12]

plt.plot(weekdays, daily_sales)
plt.title("Bakery sales this week")

plt.xlabel('Day of the week')
plt.ylabel('Sales per day')

plt.show()
```

```python
months = [1,2,3,4]
monthly_sales = [1000,1500,1200,1800]

plt.plot(months, monthly_sales, color='blue', linestyle='--', linewidth=2, marker='o', label='Monthly sales data')
plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('Monthly sales data report')

plt.legend(loc='upper left', fontsize=12)
plt.show()
```

```python
# additional attributes
months = [1,2,3,4]
monthly_sales = [1000,1500,1200,1800]

plt.plot(months, monthly_sales, color='purple', linestyle='--', linewidth=2, marker='o', label='Monthly sales data')
plt.xlim(1,5)
plt.ylim(0,2000)
plt.grid(color='pink', linestyle="-.", linewidth=2)
plt.show()
```

```python
# additional attributes
months = [1,2,3,4]
monthly_sales = [1000,1500,1200,1800]

plt.plot(months, monthly_sales, color='brown', linestyle='--', linewidth=2, marker='o', label='Monthly sales data')
plt.xticks([1,2,3,4,5], ['M1','M2','M3','M4','M5'])
plt.show()
```

# Column Chart

```python
product = ['A','B','C','D']
product_sales = [1000,1500,800,1200]

plt.bar(product, product_sales, color='orange', label='Sales 2025', width=0.5)                      # determines bar chart to be created
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()
plt.show()
```

# Pie Chart

```python
regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['gold','skyblue','lightgreen','coral'])
plt.title('Revenue contribution by region')

plt.show()
```

# Histogram Chart

```python
scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]

plt.hist(scores, bins=5, color='lightblue', edgecolor='black', width=7)
plt.xlabel('Score Range')
plt.ylabel('No. of students')
plt.title('Score distribution of students')
plt.show()
```

# Scatter Plot

```python
hours_studied = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

plt.scatter(hours_studied, exam_scores, color='purple', marker='^', label='student data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Study Time vs Marks Scored')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# grouping the data seperately
study_hrs = [1,2,3,4,5,6,7,8]
exam_scores_divA = [50,55,60,65,70,75,80,85]
exam_scores_divB = [30,35,40,45,50,55,60,65]

plt.scatter(study_hrs, exam_scores_divA, color='blue', label='divA')
plt.scatter(study_hrs, exam_scores_divB, color='yellow', label='divB')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Study Time vs Marks Scored')
plt.xlim(0,10)
plt.ylim(0,100)
plt.legend()
plt.grid(True)
plt.show()
```

# Subplot

```python
plt.subplot(2,2,1)                  # the figure has 2 row, 2 columns, and this the first plot.
plt.plot(x_vals,y_vals)
plt.title('Line Chart')

plt.subplot(2,2,2)                  # the figure has 2 row, 2 columns, and this the second plot.
plt.bar(x_vals,y_vals)
plt.title('Bar Chart')

plt.subplot(2,2,3)                  # the figure has 2 row, 2 columns, and this the third plot.
plt.pie(x_vals,labels=y_vals)
plt.title('Pie Chart')

plt.subplot(2,2,4)                  # the figure has 2 row, 2 columns, and this the fourth plot.
plt.scatter(x_vals,y_vals)
plt.title('Scatter Chart')

plt.tight_layout()                  # autofitting size of each charts in the canvas
plt.show()
```

## object oriented approach

```python
# additional attributes
fig, ax = plt.subplots(1, 2, figsize=(10,5))

ax[0].plot(x_vals,y_vals, color='blue')     # individually modifying the 1st chart
ax[0].set_title('Line Chart')

ax[1].bar(x_vals,y_vals, color='green')     # individually modifying the 2nd chart
ax[1].set_title('Bar Chart')

fig.suptitle('Line vs Bar chart')           # changes on entire canvas
plt.tight_layout()
plt.show()
```
