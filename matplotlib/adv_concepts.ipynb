```python
import matplotlib.pyplot as plt
import numpy as np
```

Plotting graph of eqn : y = x^2

```python
plt.plot([-3, -2, -1, 0, 1, 2, 3], [9, 4, 1, 0, 1, 4, 9])
plt.show()
```

```python
# to smothen the above rigged graph
x = np.arange(-3, 3, step = 0.1)
y = x ** 2                                  # to plot the square for each of x values

plt.plot(x, y)
plt.xticks(np.arange(-3, 3, step = 0.3), rotation = 90)    # x values in 0.3 intervals
plt.title("Squares") 
plt.xlabel("x-values")
plt.ylabel("square values")

plt.show()
```

```python
# cosmetic formatting

x = np.arange(-3, 3, step = 0.1)

y = x ** 2

plt.title("Squares")
plt.xlabel("x-values")
plt.ylabel("square values")

plt.plot(x, y, 'r*')  # new - *, o, ., v, -, -- for different styles
plt.show()
```

```python
# plotting with x & y range limits

x = np.arange(-3, 3, step = 0.1)

y = x ** 2

plt.title("Squares")
plt.xlabel("x-values")
plt.ylabel("square values")

plt.xlim(left = 0, right = 3) 

plt.plot(x, y, color='green')
plt.show()
```

Plotting multiple lines in one chart

```python
x = np.arange(-3, 3, step = 0.1)

y1 = x ** 2

y2 = x ** 3

plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
```

```python
# with labels and legends

x = np.arange(-3, 3, step = 0.1)
y1 = x ** 2
y2 = x ** 3

plt.plot(x, y1, label = "Squares")
plt.plot(x, y2, label = "Cubes")

plt.legend()

plt.show()
```

```python
# lengends formatting alternate way with specific location of legend
x = np.arange(-3, 3, step = 0.1)
y1 = x ** 2
y2 = x ** 3

plt.plot(x, y1)
plt.plot(x, y2)

plt.legend(["Squares", "Cubes"], loc = "lower right", ncol = 2, frameon = False)

plt.text(0, 0, "(0, 0)")
plt.text(1, 1, "(1, 1)")

plt.grid() 

plt.show()
```

```python
# shorthand syntax for plotting multiple lines
plt.plot(x, y1, x, y2)

plt.legend(["Squares", "Cubes"])

plt.text(0, 0, "(0, 0)")
plt.text(1, 1, "(1, 1)")

plt.show()
```

Scatter plot

```python
x = np.arange(0, 20)
y = np.linspace(1, 20, dtype=np.int32, num = 20) + np.random.rand(20)*10

plt.scatter(x, y)
plt.show()
```

```python
# scatter plot with multiple values
x2 = np.arange(0, 20) # new
y2 = np.linspace(1, 20, dtype=np.int32, num = 20) + np.random.rand(20)*10 # new

plt.scatter(x, y)
plt.scatter(x2, y2)

plt.show()
```

```python
# alternate way
plt.scatter(x, y, color='b')            # specifying colours explicitly
plt.scatter(x2, y2, color='r')
plt.legend(["Shop1", "Shop2"])

plt.show()
```

Scatter plot without scatter function

```python
plt.plot(x, y, 'bo')                        # plotting without scatter function cannot help with tiny customization
plt.plot(x2, y2, 'ro')                      # but its useful for large datasets
plt.legend(["Shop1", "Shop2"])

plt.show()
```

```python
plt.scatter(x, y, color = 'b', s = 4*abs(y))                    # scatter function to determine size of dots
plt.scatter(x2, y2, color = 'r', s = 4*abs(y2))
plt.legend(["Shop1", "Shop2"])

plt.show()
```

Pie Charts

###  plotting graph for sigmoid function, which is defined as:
$f(x) = \frac{1}{1+e^{-x}}$

```python
import math
import numpy as np

sigmoid = lambda x : 1/(1 + math.exp(-x))        # selecting a very broad range of x-values

x = np.arange(-100, 100)

y = [sigmoid(i) for i in x]                      # sigmoid function is scalar (not vectorized) Thus using list comprehension

plt.plot(x, y)
plt.show()
```

```python
# reducing x value range

import math
import numpy as np

sigmoid = lambda x : 1/(1 + math.exp(-x))

x = np.arange(-10, 10)
y = [sigmoid(i) for i in x]

plt.plot(x, y)
plt.show()
```

```python
# plotting a sine function

x = np.arange(-10, 10, 0.1)
y = np.sin(x)

fig = plt.plot(x, y)

plt.show()
```

```python
# plotting a tan function

x = np.arange(-10, 10, 0.1)
y = np.tan(x)

fig = plt.plot(x, y)

plt.show()
```

```python
# plotting a log function

x = np.arange(-10, 10, 0.1)
y = np.log(x)

fig = plt.plot(x, y)

plt.show()
```
