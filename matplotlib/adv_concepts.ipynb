```python
import matplotlib.pyplot as plt
import numpy as np
```

Plotting graph of eqn : y = x^2

```python
plt.plot([-3, -2, -1, 0, 1, 2, 3], [9, 4, 1, 0, 1, 4, 9])
plt.show()
```

```python
# to smothen the above rigged graph
x = np.arange(-3, 3, step = 0.1)
y = x ** 2                                  # to plot the square for each of x values

plt.plot(x, y)
plt.xticks(np.arange(-3, 3, step = 0.3), rotation = 90)    # x values in 0.3 intervals
plt.title("Squares") 
plt.xlabel("x-values")
plt.ylabel("square values")

plt.show()
```

```python
# cosmetic formatting

x = np.arange(-3, 3, step = 0.1)

y = x ** 2

plt.title("Squares")
plt.xlabel("x-values")
plt.ylabel("square values")

plt.plot(x, y, 'r*')  # new - *, o, ., v, -, -- for different styles
plt.show()
```

```python
# plotting with x & y range limits

x = np.arange(-3, 3, step = 0.1)

y = x ** 2

plt.title("Squares")
plt.xlabel("x-values")
plt.ylabel("square values")

plt.xlim(left = 0, right = 3) 

plt.plot(x, y, color='green')
plt.show()
```

Plotting multiple lines in one chart

```python
x = np.arange(-3, 3, step = 0.1)

y1 = x ** 2

y2 = x ** 3

plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
```

```python
# with labels and legends

x = np.arange(-3, 3, step = 0.1)
y1 = x ** 2
y2 = x ** 3

plt.plot(x, y1, label = "Squares")
plt.plot(x, y2, label = "Cubes")

plt.legend()

plt.show()
```

```python
# lengends formatting alternate way with specific location of legend
x = np.arange(-3, 3, step = 0.1)
y1 = x ** 2
y2 = x ** 3

plt.plot(x, y1)
plt.plot(x, y2)

plt.legend(["Squares", "Cubes"], loc = "lower right", ncol = 2, frameon = False)

plt.text(0, 0, "(0, 0)")
plt.text(1, 1, "(1, 1)")

plt.grid() 

plt.show()
```

```python
# shorthand syntax for plotting multiple lines
plt.plot(x, y1, x, y2)

plt.legend(["Squares", "Cubes"])

plt.text(0, 0, "(0, 0)")
plt.text(1, 1, "(1, 1)")

plt.show()
```

Multiple Bar chart

```python
plt.bar([1, 2, 3], [10, 20, 5], width=0.2, label='Market1')
plt.bar(np.array([1, 2, 3])+0.2, [5, 10, 5], width=0.2, label='Market2') # new
plt.legend()
plt.show()
```

Histogram with random values

```python
# same code
vals = np.random.rand(100)
plt.hist(vals, 20) # Let's change the no. of bins to 20
plt.show()
```

```python
# histogram with explicit bin size setting

vals = np.random.rand(100)
#plt.hist(vals, 20)
plt.hist(vals, [0,0.5,0.7,0.9,1])
plt.show()
```

Scatter plot

```python
x = np.arange(0, 20)
y = np.linspace(1, 20, dtype=np.int32, num = 20) + np.random.rand(20)*10

plt.scatter(x, y)
plt.show()
```

```python
# scatter plot with multiple values
x2 = np.arange(0, 20) # new
y2 = np.linspace(1, 20, dtype=np.int32, num = 20) + np.random.rand(20)*10 # new

plt.scatter(x, y)
plt.scatter(x2, y2)

plt.show()
```

```python
# alternate way
plt.scatter(x, y, color='b')
plt.scatter(x2, y2, color='r')
plt.legend(["Shop1", "Shop2"])

plt.show()
```
