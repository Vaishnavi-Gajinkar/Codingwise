## GOAL - To check if coin is biased based on number of Heads appearing

Our null hypothesis (Ho) is that "Coin is fair" <br>
Hence alternate hypothesis (Ha) automatically is that "Coin is biased"

### CASE 1 - Number of tosses = 10

```python
# importing required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot
```

```python
num_tosses = 10
dist = stats.binom(num_tosses, 0.5)
```

Both probability mass function (pmf) and cumulative distribution function (cdf) can help find the 'p' value

```python
# PMF

dist.pmf(7) + dist.pmf(8) + dist.pmf(9) + dist.pmf(10)              # Prob[T >= 7]
```

```python
# CDF

1 - dist.cdf(6)                                         # 1 - Prob[T <= 6]
```

Since p=0.17 i.e. 17% > 5%<br>
We can accept Ho and reject Ha

```python
# Plotting p values uptill the count of total tosses
x_values = np.arange(0, num_tosses+1)
y_values = dist.pmf(x_values)
plt.bar(x_values, y_values)
```

From graph its clear that our null hypothesis can be accepted for count of tosses between 2 & 8. <br>
after and before which p<5% hence Ha will be accepted

### CASE 2 - Number of tosses = 100

```python
num_tosses = 100
dist = stats.binom(num_tosses, 0.5) 
1 - dist.cdf(69)
```

Comparing with previous graph, its seen that Y axis peak has reduced from 0.25 to 0.08

### CASE 3 - Number of tosses = 1000

```python
num_tosses = 1000
dist = stats.binom(num_tosses, 0.5)
1 - dist.cdf(699)
```

```python
x_values = np.arange(0, num_tosses+1)
y_values = dist.pmf(x_values)
plt.bar(x_values, y_values)
```

Further increasing the number of tosses, has reduced the peak of P value down till 0.025.<br>
Since the max p value is less than 5%, the alternate hypothesis has to be accepted here.
